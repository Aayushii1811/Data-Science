# PLOTS AND SUBPLOTS

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

```python
ax1.plot(x,y,'b')
ax2.plot(x,y,'b')
fig
```

# SUBPLOTS(OBJECT-ORIENTED METHOD)

```python
fig,axes=plt.subplots(2,2)

axes[0,0].plot(x,y,'r')
axes[0,0].set_xlabel('x')
axes[0,0].set_ylabel('y')
axes[0,0].set_title('title');

axes[0,1].plot(x,y,'b')
axes[0,1].set_xlabel('x')
axes[0,1].set_ylabel('y')
axes[0,1].set_title('title');

axes[1,0].plot(x,y,'g')
axes[1,0].set_xlabel('x')
axes[1,0].set_ylabel('y')
axes[1,0].set_title('title');

axes[1,1].plot(x,y,'m')
axes[1,1].set_xlabel('x')
axes[1,1].set_ylabel('y')
axes[1,1].set_title('title');
```

```python
import pandas as pd
```

```python
f=[['Aayushi',24,26,28,30],
   ['Kavya',27,21,21,12],
   ['Shriya',23,21,12,15]]
df=pd.DataFrame(f,columns=['Name','Biology','Chemistry','Maths','Physics'])
df
```

```python
fig,axes=plt.subplots(2,2,layout="constrained")

axes[0,0].plot(df.Name,df.Biology,'r',label='Biology')
axes[0,0].set_xlabel('Name')
axes[0,0].set_ylabel('Marks')
axes[0,0].legend()


axes[0,1].plot(df.Name,df.Chemistry,'g',label='Chemistry')
axes[0,1].set_xlabel('Name')
axes[0,1].set_ylabel('Marks')
axes[0,1].legend()

axes[1,0].plot(df.Name,df.Maths,'b',label='Maths')
axes[1,0].set_xlabel('Name')
axes[1,0].set_ylabel('Marks')
axes[1,0].legend()


axes[1,1].plot(df.Name,df.Physics,'m',label='Physics')
axes[1,1].set_xlabel('Name')
axes[1,1].set_ylabel('Marks')
axes[1,1].legend()
```
